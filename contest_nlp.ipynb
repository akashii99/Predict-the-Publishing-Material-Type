```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

import re
import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')

df = pd.read_csv('train_file.csv')
print(df.head())
print(df.info())
print(df['Subjects'].nunique())
print(df['Creator'].nunique())
print(df['Creator'].nunique())
```

```python
df2 = pd.read_csv('test_file.csv')
print(df2.head())
```

```python
df['Subjects'].fillna("Others", inplace=True)
#print(df.info())
df['Publisher'].fillna("Others", inplace=True)
#print(df.info())

df['Creator'].fillna("Single", inplace=True)
df['PublicationYear'].fillna("Not_Mentioned", inplace=True)

df['Checkouts'] = df['Checkouts'].astype(str)
print(df.info())

X = df.Subjects + df.Publisher + df.Checkouts + df.PublicationYear
Y = df['MaterialType']
#print(Y)
```

Cleaning text data

```python
def clean_text(text):
    text = text.str.strip().str.lower()
    #text = text.str.translate(str.maketrans('', '', '():,.?\'"/+-_[]&%$@'))
    
    return text
```

```python
X = clean_text(X)
```

```python
vectorizer = CountVectorizer(min_df=1)
vec = vectorizer.fit_transform(X)

words = vectorizer.get_feature_names()
print(words)
words_counts = vec.toarray()
#print(words_counts)

transformer = TfidfTransformer()
tfidf = transformer.fit_transform(words_counts)
#print(tfidf.toarray())
```

```python
vectors = tfidf
```

```python
clf = MultinomialNB(alpha=.18,class_prior=None, fit_prior=True)

#  Train the classifier
clf.fit(vectors, Y)
```

```python
df2['Subjects'].fillna("Others", inplace=True)
print(df2.info())

df2['Publisher'].fillna("Others", inplace=True)

df2['Creator'].fillna("Others", inplace=True)
df2['PublicationYear'].fillna("Not_Mentioned", inplace=True)
df2['Checkouts'] = df2['Checkouts'].astype(str)
print(df.info())

X1 = df2.Subjects + df2.Publisher + df2.Checkouts + df2.PublicationYear
```

```python
X1 = clean_text(X1)
```

```python
vectors_test = vectorizer.transform(X1)
```

```python
pred = clf.predict(vectors_test)
print(pred)
```


Storing the predictions of MaterialType and Saving to csv file



```python
df3 = pd.DataFrame()
df3['ID'] = df2['ID']
df3['MaterialType'] = pred

print(df3.head())

df3.to_csv('Submission.csv', index=False)
```
